```python
import numpy as np
from scipy.spatial.distance import squareform
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# 示例数据，假设 B 是已经计算好的数据点之间的距离矩阵
B = np.array([
    [0, 0.5, 0.8, 0.2],
    [0.5, 0, 0.7, 0.3],
    [0.8, 0.7, 0, 0.6],
    [0.2, 0.3, 0.6, 0]
])

# 将距离矩阵转换为适用于 linkage 函数的形式
condensed_distance = squareform(B)

# 构建层次聚类树
linkage_matrix = linkage(condensed_distance, method='average')

# 绘制树谱图
dendrogram(linkage_matrix, labels=list(range(len(B))), leaf_rotation=90, leaf_font_size=8, orientation='left', color_threshold=np.sqrt(5) + 0.1)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Distance')
plt.ylabel('Data Points')
plt.show()

linkage_matrix
```
